#What Are Databricks Widgets?
Databricks widgets are UI elements embedded in notebooks that allow users to input values dynamically. They make notebooks more flexible and reusable by enabling parameterization.


##Create a widget (Python example):

```python
dbutils.fs.help()
```

```python
dbutils.widgets.help()
```

```python
#Create Text box
dbutils.widgets.text("customer_id", "12345", "Enter Customer ID")
```

```python
#capture the widget input in a variable
Customer_ID=dbutils.widgets.get("customer_id")
#use that variable for some purpose
print(f"Given Customer is  {Customer_ID}")
```

###Combobox widget - Used to choose only one value from the dropdown by searching

```python

dbutils.widgets.combobox("country_combo","India",["India","USA","UK","Canada","Australia"])
country=dbutils.widgets.get("country_combo")
print(f"Country is {country}")
```

###Multiselect widget  - Used to choose multiple values from the dropdown by searching

```python
dbutils.widgets.multiselect("hobbies_multiselect","Dance",["Dance","Music","Sports","Reading","Writing"])
hobbies=dbutils.widgets.get("hobbies_multiselect")
print(f"Hobbies are {hobbies}",type(hobbies))
print("Top and Least hobbies ?", hobbies.split(",")[0],hobbies.split(",")[-1])
```

#### Imporantant interview question: Difference between run magic and dbutils command?
##### MAGIC A. The %run magic command will run some other notebook inline in this current notebook itself, so we don't have write some other notebook code, rather we can just run it...<br>
##### MAGIC B. The dbutils.notebook.run() command will trigger some other notebook in the respective notebook environment itself, and we can add additional parameters such as timeout seconds and custom parameters to the widgets we can pass...


```python
#%run "/Workspace/Users/viggneshwar@gmail.com/databricks/Databricks_Sample/markdown"

return_status=dbutils.notebook.run("/Workspace/Users/viggneshwar@gmail.com/databricks/Databricks_Sample/parameter_notebook",90,{"table_name":"cities"})
print("child notebook ",return_status)

```
